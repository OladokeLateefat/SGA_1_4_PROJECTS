```python
# # query of customers that did not place any orders

sql = """
SELECT name
FROM(SELECT customers.name, orders.order_id
FROM customers 
INNER JOIN orders 
ON customers.id = orders.id)
WHERE order_id NOT IN (SELECT order_id FROM orders)
"""

# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table
```

### b. Write a query that determines the highest earner in each department

```python
#  query that determines the highest earner in each department¶

sql = """
SELECT name, dept_name, ROUND(MAX(salary),2) AS max_salary
FROM (SELECT employee.name, employee.salary, department.dept_name
FROM employee 
INNER JOIN department 
ON employee.dept_id = department.id
ORDER BY dept_name)
GROUP BY dept_name
"""

# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table


# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table
```

### c. Write a query that shows all the employees in each department except the lowest earner. i.e generate the table below

```python
#  query that determines the lowest earner in each department¶

sql = """
SELECT name,dept_name, ROUND(MIN(salary),2) AS min_salary
FROM (SELECT employee.name, employee.salary, department.dept_name
FROM employee 
INNER JOIN department 
ON employee.dept_id = department.id
ORDER BY dept_name)
GROUP BY dept_name
"""


# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table


# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table
```

```python
sql = """SELECT name, dept_name, salary
FROM(SELECT employee.name, employee.salary, department.dept_name
FROM employee 
INNER JOIN department 
ON employee.dept_id = department.id
ORDER BY dept_name)
GROUP BY dept_name
HAVING (SELECT  dept_name, ROUND(MIN(salary),2) AS min_salary
FROM (SELECT employee.name, employee.salary, department.dept_name
FROM employee 
INNER JOIN department 
ON employee.dept_id = department.id
ORDER BY dept_name)) != (SELECT dept_name, ROUND(MAX(salary),2) AS max_salary
FROM (SELECT employee.name, employee.salary, department.dept_name
FROM employee 
INNER JOIN department 
ON employee.dept_id = department.id
ORDER BY dept_name))
"""
# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table
```

### b. Write a query that selects ONLY the students that scored more than the mean of the percentage_score.


```python
# viewing the students  table

sql = """
SELECT  name,gender, SUM((maths+english+physics)/3) *100 AS percentage_score
FROM students
GROUP BY matric_no
ORDER BY percentage_score DESC ;
"""

# execute the query
cursor.execute(sql)

# execute the query
table = pd.read_sql(sql,conn)
table
```
